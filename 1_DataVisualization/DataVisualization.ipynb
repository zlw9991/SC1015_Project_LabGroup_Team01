```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

sb.set()

table = pd.read_csv('cmb1.csv')
f = plt.figure(figsize=(12, 12))
sb.heatmap(table.corr(), vmin = -1, vmax = 1, annot = True, fmt = ".2f")
```

```python
# we can see it is the previous day's MKTCP (market cap % change) that has the strongest correlation with the current days
# MKPRU (market price % change)
# this is followed LTCPCUSD
```

```python
# Further visualization confirms this
print(table.head())
mkpru = table.MKPRU
mktcp = table.MKTCP
ltc = table.LTCPCUSD
```

```python
import pandas as pd
  
# importing matplotlib module
import matplotlib.pyplot as plt
plt.style.use('default')
  
# %matplotlib inline: only draw static
# images in the notebook
%matplotlib inline
plt.figure(figsize=(16, 8), dpi=150)
  
# using plot method to plot open prices.
# in plot method we set the label and color of the curve.
mkpru.plot(label='MKPRU', color='red')
#mktcp.plot(label='MKTCP')
#ltc.plot(label='LTCPCUSD')
  
# adding title to the plot
plt.title('Percentage Change')
  
# adding Label to the x-axis
plt.xlabel('Days')
  
# adding legend to the curve
plt.legend()


plt.figure(figsize=(16, 8), dpi=150)
mktcp.plot(label='MKTCP', color='blue')
#mktcp.plot(label='MKTCP')
#ltc.plot(label='LTCPCUSD')
  
# adding title to the plot
plt.title('Percentage Change')
  
# adding Label to the x-axis
plt.xlabel('Days')
  
# adding legend to the curve
plt.legend()

plt.figure(figsize=(16, 8), dpi=150)
ltc.plot(label='LTCPCUSD', color='orange')
#mktcp.plot(label='MKTCP')
#ltc.plot(label='LTCPCUSD')
  
# adding title to the plot
plt.title('Percentage Change')
  
# adding Label to the x-axis
plt.xlabel('Days')
  
# adding legend to the curve
plt.legend()
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set()
#numDF = pd.DataFrame(table[["MKPRU","MKTCP","LTCPCUSD"]])
numDF = pd.DataFrame(table[["MKPRU","LTCPCUSD"]])


plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
sb.lmplot(x='MKPRU', y='LTCPCUSD', data=numDF)
plt.title("Scatter Plot with Linear fit");
```
